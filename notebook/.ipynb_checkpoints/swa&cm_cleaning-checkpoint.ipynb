```python
import pandas as pd
import numpy as np 
```

```python
df1=pd.read_csv(r"C:\Users\user\Downloads\safe water\who_unicef_jmp__urban_rural.csv")
df1
```

```python
df2=pd.read_csv(r"C:\Users\user\Downloads\safe water\worldbank_wdi_urban_rural.csv")
df2
```

```python
# Assuming df1 and df2 are already defined

# Perform full outer merge on shared columns
merged_df = pd.merge(
    df1,
    df2,
    on=["Country", "Region", "IncomeGroup", "AreaType", "Year"],
    how="outer"  # keeps all rows from both df1 and df2
)

# Save the merged result
merged_df.to_csv("merged_full_dataset.csv", index=False)

# Optional: Display summary
print(" Merged dataset shape:", merged_df.shape)
print(" Columns:", list(merged_df.columns))
print("Preview:")
print(merged_df.head())
```
